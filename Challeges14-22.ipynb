```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ดึงข้อมูลทั้งหมดของหุ้น NVIDIA
nvda = yf.Ticker("NVDA")

# ดึงข้อมูลหุ้นทั้งหมด
nvda_history = nvda.history(period="max")

# ดูข้อมูล 5 แถวแรก
nvda_history.head()
```

```python
# ดูข้อมูลทั่วไปเกี่ยวกับ NVIDIA
nvda.info
```

```python
# ข้อมูลปัจจุบันของหุ้น
nvda.history(period="1d")
```

```python
# ดูข้อมูล valuation และ financials ของ NVIDIA
nvda.financials
nvda.balance_sheet
nvda.cashflow
```

```python
nvda.recommendations
```

```python
plt.figure(figsize=(12,6))
plt.plot(nvda_history.index, nvda_history['Close'], label="NVIDIA Stock Price", color='blue')
plt.title("NVIDIA Stock Price Over Time")
plt.xlabel("Year")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.grid()
plt.show()
```

```python
nvda_history['SMA50'] = nvda_history['Close'].rolling(window=50).mean()
nvda_history['SMA200'] = nvda_history['Close'].rolling(window=200).mean()

plt.figure(figsize=(12,6))
plt.plot(nvda_history.index, nvda_history['Close'], label="NVIDIA Stock Price", color='blue')
plt.plot(nvda_history.index, nvda_history['SMA50'], label="50-day SMA", color='red')
plt.plot(nvda_history.index, nvda_history['SMA200'], label="200-day SMA", color='green')
plt.title("NVIDIA Stock Price with Moving Averages")
plt.xlabel("Year")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.grid()
plt.show()
```

```python
nvda_history.to_csv("NVIDIA_stock_data.csv")
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# โหลดข้อมูล Iris จาก scikit-learn
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
# ดูข้อมูลเบื้องต้น
iris_df.head()  # ดูแค่ 5 แถวแรก
iris_df.info()  # ดูข้อมูลประเภท และจำนวนข้อมูล
iris_df.describe()  # สถิติพื้นฐาน เช่น ค่าเฉลี่ย, ค่าเบี่ยงเบนมาตรฐาน
```

```python
iris_df.isnull().sum()  # ตรวจสอบค่าว่างในแต่ละคอลัมน์
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# การสร้าง Histogram สำหรับคอลัมน์ต่างๆ
sns.histplot(iris_df['sepal length (cm)'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
sns.pairplot(iris_df, hue='target')  # ดูความสัมพันธ์ระหว่างฟีเจอร์ พร้อมแยกตาม target
plt.show()
```

```python
correlation = iris_df.corr()  # คำนวณความสัมพันธ์
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')  # แสดงผล heatmap
plt.show()
```

```python
sns.boxplot(x='target', y='sepal length (cm)', data=iris_df)
plt.title('Boxplot of Sepal Length')
plt.show()
```

```python
import yfinance as yf

# ดึงข้อมูลหุ้น NVIDIA ตั้งแต่ปี 2020 ถึง 2025
nvidia_data = yf.download("NVDA", start="2020-01-01", end="2025-01-01")
nvidia_data.head()  # ดูข้อมูลเบื้องต้น
```

```python
nvidia_data.info()  # ดูข้อมูลประเภทและจำนวนข้อมูล
nvidia_data.describe()  # ดูสถิติพื้นฐาน
```

```python
nvidia_data.isnull().sum()  # ตรวจสอบค่าว่างในข้อมูล
```

```python
sns.histplot(nvidia_data['Close'], kde=True)
plt.title('Distribution of NVIDIA Closing Price')
plt.show()
```

```python
nvidia_data['Close'].plot(title='NVIDIA Stock Price Over Time', figsize=(10, 6))
plt.show()
```

```python
correlation = nvidia_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of NVIDIA Stock Data')
plt.show()
```

```python
nvidia_data['MA50'] = nvidia_data['Close'].rolling(window=50).mean()  # ค่าเฉลี่ย 50 วัน
nvidia_data['MA200'] = nvidia_data['Close'].rolling(window=200).mean()  # ค่าเฉลี่ย 200 วัน

# แสดงกราฟราคาและ Moving Average
nvidia_data[['Close', 'MA50', 'MA200']].plot(figsize=(10, 6))
plt.title('NVIDIA Stock Price with Moving Averages')
plt.show()
```

```python
nvidia_data['Year'] = nvidia_data.index.year
nvidia_data.groupby('Year')['Close'].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Average NVIDIA Stock Price by Year')
plt.show()
```
